**Volatility Estimators — Historical Standard Deviation**

#### Log-Returns

All volatility estimation operates on daily **continuously compounded log-returns**:

$$r_t = \ln\!\left(\frac{S_t}{S_{t-1}}\right)$$

#### Full-Sample Historical Standard Deviation (Part 1 & Part 3)

The full-sample annualised volatility is:

$$\hat{\sigma}_{\text{hist}} = \sqrt{252} \cdot \operatorname{std}(r_1, r_2, \ldots, r_T)$$

where $T = 865$ daily log-returns (866 trading days, 02 Jan 2023 – 01 Jun 2026). This single constant is used as `SIGMA_CPN` in the Part 1 at-par design.

#### 30-Day Rolling Historical Volatility (Part 2 Pricing, Tasks V–VII)

For the daily certificate valuation, a **time-varying** estimate is used:

$$\hat{\sigma}_t^{(30)} = \sqrt{252} \cdot \operatorname{std}\!\left(r_{t-29}, \ldots, r_t\right)$$

On each valuation day $t$, the 30 most recent daily log-returns are used to form an annualised vol estimate. This is looked up via `rolling_vol_30.asof(date)` in the pricing loop, so the binomial tree captures the prevailing volatility regime on each date rather than a fixed full-period average.

#### Why the estimator split?

The full-sample estimator is appropriate for design, where a stable long-run vol is needed. The rolling estimator is appropriate for daily mark-to-model pricing, where a contemporaneous risk measure is more relevant — the 30-day window captures the current volatility regime without overweighting the entire 2023–2026 history. The Part 3 Monte-Carlo uses a 252-day trailing estimate that balances the two: recent enough to reflect the prevailing regime, long enough to be stable over a one-year simulation horizon.

### Risk-Free Rate: Svensson (1994) Yield-Curve Calibration

Implements the Nelson-Siegel-Svensson parametric spot-rate model that supplies every risk-free rate in this notebook. Daily ECB AAA sovereign yield-curve parameters (β₀ β₁ β₂ β₃ τ₁ τ₂) are loaded from `data/ECB Data Portal_20260615151519.csv` — downloaded from the [ECB Data Portal – AAA yield curve parameters](https://data.ecb.europa.eu/data/data-categories/financial-markets-and-interest-rates/euro-area-yield-curves/aaa-rated-government-bonds-yield-curve/parameters). Only dates from **04 Sep 2025** onward (the DU2076 valuation window) are used. On each valuation day the six parameters for that date are retrieved and the Svensson formula is evaluated at the **remaining maturity** of the certificate on that day, `r(T_rem)`. This gives a **daily, maturity-matched** risk-free rate rather than a single fixed benchmark. A static `SV_P1` snapshot (01 Jun 2026 parameters) is retained for the Part 1 product-design pricing (3-year spot rate), for the constant 10-year benchmark `RISK_FREE` used in the Task VI–VII Greeks/replication and in Part 3, and as a fallback for any missing dates.


````python
# ══════════════════════════════════════════════════════════════════════════════
# SVENSSON (1994) RISK-FREE RATE — supplies all risk-free rates in this notebook
# ECB AAA sovereign yield curve — daily parameters from ECB Data Portal
# Source: https://data.ecb.europa.eu/data/data-categories/financial-markets-and-interest-rates/euro-area-yield-curves/aaa-rated-government-bonds-yield-curve/parameters
# ══════════════════════════════════════════════════════════════════════════════

def svensson(t, b0, b1, b2, b3, tau1, tau2):
    # Svensson (1994) continuously compounded spot rate for maturity t (years).
    f1 = (1 - np.exp(-t / tau1)) / (t / tau1)
    f2 = f1 - np.exp(-t / tau1)
    f3 = (1 - np.exp(-t / tau2)) / (t / tau2) - np.exp(-t / tau2)
    return (b0 + b1*f1 + b2*f2 + b3*f3) / 100

# ── Static snapshot (01 Jun 2026) — used for Part 1 design pricing ───────────
SV_P1 = dict(b0=1.330382, b1=0.700052,  b2=1.816755,
             b3=7.130184, tau1=0.966763, tau2=15.839369)

MATURITY_DATE = pd.Timestamp('2027-06-18')
RISK_FREE     = svensson(10.0, **SV_P1)   # 10y benchmark: Tasks VI-VII & Part 3

# ── Daily Svensson parameter lookup — ECB Data Portal download ────────────────
# File covers 2004–2026; we filter to the DU2076 valuation window (Sep 2025+).
_sv_raw = pd.read_csv(
    'data/ECB Data Portal_20260615151519.csv',
    usecols=[0, 2, 3, 4, 5, 6, 7],     # DATE + 6 parameter columns
    header=0,
)
_sv_raw.columns = ['date', 'b0', 'b1', 'b2', 'b3', 'tau1', 'tau2']
_sv_raw['date'] = pd.to_datetime(_sv_raw['date'])
_sv_daily = (
    _sv_raw[_sv_raw['date'] >= '2025-09-01']
    .set_index('date')
    .apply(pd.to_numeric, errors='coerce')
    .sort_index()
)
_sv_index = _sv_daily.index

def get_risk_free_rate(date, T_rem):
    # Svensson spot rate for remaining maturity T_rem on a given date.
    #
    # Forward-fills: uses the most recent available ECB business day on or before
    # `date`. Falls back to SV_P1 if date is before the downloaded series.
    idx = _sv_index.searchsorted(date, side='right') - 1
    if idx < 0:
        return svensson(T_rem, **SV_P1)
    row = _sv_daily.iloc[idx]
    return svensson(T_rem, row['b0'], row['b1'], row['b2'],
                    row['b3'], row['tau1'], row['tau2'])

print(f'ECB Svensson 10y benchmark (01 Jun 2026, Tasks VI-VII & Part 3): r(10y) = {RISK_FREE:.4%}')
print(f'Daily Svensson params loaded: {len(_sv_daily)} dates '
      f'({_sv_index[0].date()} → {_sv_index[-1].date()})')
_d0, _d1 = pd.Timestamp('2025-09-04'), pd.Timestamp('2026-06-01')
_T0 = max((MATURITY_DATE - _d0).days / 365, 1/365)
_T1 = max((MATURITY_DATE - _d1).days / 365, 1/365)
print(f'r(T_rem={_T0:.2f}y) on {_d0.date()} = {get_risk_free_rate(_d0, _T0):.4%}')
print(f'r(T_rem={_T1:.2f}y) on {_d1.date()} = {get_risk_free_rate(_d1, _T1):.4%}')
````

**Svensson (1994) Parameter Source**

The six Nelson-Siegel-Svensson parameters (`β₀ β₁ β₂ β₃ τ₁ τ₂`) are the daily estimates published by the **ECB** for the AAA-rated euro area sovereign yield curve, downloaded from the [ECB Data Portal – AAA yield curve parameters](https://data.ecb.europa.eu/data/data-categories/financial-markets-and-interest-rates/euro-area-yield-curves/aaa-rated-government-bonds-yield-curve/parameters) and stored locally in `data/ECB Data Portal_20260615151519.csv`.

On each valuation day `get_risk_free_rate(date, T_rem)` looks up that day's parameters (forward-filling weekends/holidays with the most recent ECB business day) and evaluates `svensson(T_rem, ...)` at the certificate's **remaining maturity** on that day. This ensures both the rate level and the term-structure shape reflect prevailing market conditions throughout the valuation window.

## Task II – Product Design

### Product: Capped Capital Protected Participation Note on Rheinmetall AG (RHM.DE)

**Underlying:** Rheinmetall AG (RHM.DE), XETRA. The same underlying runs through Parts 1 and 2 for consistency — the product designed in Part 1 is directly comparable to DU2076 valued in Part 2.

**Intuition:** The core idea is to take most of the invested capital and put it in a zero-coupon bond, guaranteeing the nominal floor at maturity. Whatever is left over — the "option budget" — buys an ATM call, providing participation in Rheinmetall's upside. Selling an OTM call at 130% brings in additional premium that funds a higher participation rate than the budget alone would allow. The investor gives up extreme upside above 30%, but for someone whose central scenario is a 15–30% gain, that is not much of a concession.

What makes this trade-off work well at high volatility is that both the long ATM call and the short cap call benefit from elevated σ. The spread (long minus short) is less sensitive to vol than either individual option, so the participation rate is reasonably stable even as σ fluctuates — a useful property given Rheinmetall's vol history.

### Component Decomposition

| Component | Position | Role |
|-----------|----------|------|
| Zero-coupon bond (face = S₀) | Long | Guarantees full nominal capital return at maturity |
| ATM European call (K = S₀ = 1,207) | Long (α units) | Captures Rheinmetall upside above S₀ |
| OTM European call (K = Cap = 1,569) | **Short** (α units) | Caps participation at 130% of S₀ → max product gain +25.6% (= α × 30%); premium funds higher α |

**Replication formula (at-par condition):**

```
CPN₀  =  ZCB₀  +  α × (Call_ATM − Call_Cap)  =  S₀ = 1,207
```

**Participation rate solved at par (code output, cell below):**

| Component | EUR |
|-----------|----:|
| ZCB present value | 1,115.79 |
| ATM call K=1,207 | 338.64 |
| Cap call K=1,569 | 231.59 |
| Net spread (Call_ATM − Call_Cap) | 107.04 |
| **Participation α = (1,207 − 1,115.79) / 107.04** | **85.2%** |

### Payoff at Maturity

$$\text{CPN}_T = S_0 + \alpha \cdot \bigl[\max(S_T - S_0,\, 0) - \max(S_T - \text{Cap},\, 0)\bigr]$$

- **S_T ≤ 1,207:** Floor pays €1,207 regardless (capital protected)
- **1,207 < S_T ≤ 1,569:** Pays €1,207 + 0.852 × (S_T − 1,207) — investor captures 85.2% of the gain
- **S_T > 1,569:** Capped payoff ≈ €1,516 — upside above 30% is foregone

**On the cap:** The cap at €1,569 (130% of S₀) is the right choice for this investor profile. Their central scenario is 15–30% appreciation over 3 years — outcomes fully inside the participation range. Scenarios above 130% require exceptional outperformance beyond the central case; those gains are given up in exchange for the participation rate that the short cap call makes possible.


````python
# ── Task II – Participation rate α (at-par design condition) ──────────────────
# The product issues at par (nominal = 100).  Setting C₀ = 100 and solving:
#
#   100  =  ZCB₀(r, T)  +  α × [Call(K=S₀) − Call(K=Cap)]
#   α    =  (100 − ZCB₀) / (Call(K=100) − Call(K=130))
#
# Note: Part 1 requires no pricing.  This derivation verifies α is economically
# sensible given the current market inputs; it is not a valuation exercise.

R_CPN = svensson(T_CPN, **SV_P1)   # 3-year Svensson spot rate (design date)

def _bs_call(S, K, r, q, sig, T):
    d1 = (np.log(S/K) + (r - q + 0.5*sig**2)*T) / (sig*np.sqrt(T))
    d2 = d1 - sig*np.sqrt(T)
    return S*np.exp(-q*T)*norm.cdf(d1) - K*np.exp(-r*T)*norm.cdf(d2)

_zcb0      = S0_CPN * np.exp(-R_CPN * T_CPN)
_call_long = _bs_call(S0_CPN, K_CPN,     R_CPN, Q_CPN, SIGMA_CPN, T_CPN)
_call_cap  = _bs_call(S0_CPN, K_CAP_CPN, R_CPN, Q_CPN, SIGMA_CPN, T_CPN)
_spread    = _call_long - _call_cap
ALPHA_CPN  = float(np.clip((S0_CPN - _zcb0) / _spread, 0.01, 2.0))
MAX_GAIN_CPN = ALPHA_CPN * (K_CAP_CPN - K_CPN)

print(f'3-year spot rate (Svensson):  r(3y)  = {R_CPN:.4%}')
print(f'Zero-coupon bond PV:           ZCB₀   = {_zcb0:.4f}')
print(f'ATM call  K={K_CPN:.0f}:                      = {_call_long:.4f}')
print(f'OTM cap call K={K_CAP_CPN:.0f} (short):           = {_call_cap:.4f}')
print(f'Spread (long − short):                 = {_spread:.4f}')
print(f'Participation rate:            α       = {ALPHA_CPN:.4f}  ({ALPHA_CPN:.1%})')
print(f'Maximum gain (S_T ≥ Cap):              = {MAX_GAIN_CPN:.2f} EUR  (+{MAX_GAIN_CPN/S0_CPN:.1%} of nominal)')
````

### Payoff at Maturity

$$V_T = \begin{cases} 2{,}000 & \text{if } S_t > 1{,}050 \;\forall\, t \in [\text{04 Sep 2025},\, \text{18 Jun 2027}] \\ \min(S_T,\; 2{,}000) & \text{if barrier breached at any point} \end{cases}$$

**Intuition:** As long as Rheinmetall never closes at or below €1,050 (≈ 61% of the issue-date spot of ~€1,708, and ≈ 13% below the 01 Jun 2026 underlying level of €1,207.00) during the product's life, the investor receives the full **€2,000 bonus** regardless of where the stock ends up — a **+61.8% return** relative to the 01 Jun 2026 certificate price of €1,236.07. If the barrier is breached, the product converts into a capped linear position in the stock: the investor receives whatever the stock is worth at maturity, capped at €2,000. The certificate thus exchanges dividend rights for conditional downside protection and a bonus above the current stock price.

## Pricing Engine

The DU2076 Bonus Cap Certificate is decomposed into two building blocks priced simultaneously on the same binomial tree:

$$V^{\text{cert}}(S_0) = \underbrace{V^{\text{DO-put}}(S_0;\; K_{\text{bonus}},\, B)}_{\text{down-and-out put}} \;+\; \underbrace{V^{\text{fwd-cap}}(S_0;\; K_{\text{cap}})}_{\min(S_T,\, K_{\text{cap}})}$$

The capped forward pays $\min(S_T, K_{\text{cap}})$ at maturity — equivalent to holding the stock and selling a call struck at $K_{\text{cap}}$. The down-and-out put pays $\max(K_{\text{bonus}} - S_T,\, 0)$ **only if** the stock never touched the barrier $B$ during the product's life.

### Why not Black-Scholes?

The standard Black-Scholes formula prices options whose payoff depends **only on the terminal stock price $S_T$**:

$$C = S_0 \cdot N(d_1) - e^{-rT} \cdot K \cdot N(d_2)$$

It models a single jump from $S_0$ to $S_T$ — the log-normal terminal distribution — and has no concept of what the stock did in between. For a vanilla call or put that is fine, since the payoff only depends on where the stock **ends up** at expiry.

The DU2076 barrier condition is fundamentally different: **if the stock touches $B = 1{,}050$ EUR at any point between today and June 2027, the bonus is permanently lost** — even if the stock fully recovers to 1,900 by maturity. This is a statement about the entire price path, not just the endpoint. Black-Scholes cannot answer *"did the stock ever cross 1,050 during its life?"* because it never modelled those intermediate steps.

This property is called **path dependency**. The binomial tree resolves it by moving in small steps $\Delta t = T/N$, checking the barrier at every node, and immediately zeroing out any branch where $S_{n,j} \leq B$. Those knocked-out branches carry zero value backwards through the entire tree, correctly reducing the certificate price to reflect the real knock-out risk. Any product whose payoff depends on the stock's behaviour *along the way* — barriers, Asian options, lookbacks — requires either a tree, Monte Carlo, or a specially derived closed-form; the basic Black-Scholes formula is insufficient.


````python
def bs_vanilla_put(S, K, r, q, sigma, T):
    # Black-Scholes European vanilla put price.
    if T <= 0:
        return float(max(K - S, 0.0))
    d1 = (np.log(S / K) + (r - q + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    return K * np.exp(-r * T) * norm.cdf(-d2) - S * np.exp(-q * T) * norm.cdf(-d1)


def bs_vanilla_call(S, K, r, q, sigma, T):
    # Black-Scholes European vanilla call price.
    if T <= 0:
        return float(max(S - K, 0.0))
    d1 = (np.log(S / K) + (r - q + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    return S * np.exp(-q * T) * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
````

### Binomial Tree — Pricing Functions

The two functions below implement the **binomial tree** used to price the DU2076 Bonus Cap Certificate. Both path-dependent components — the **down-and-out put** and the **capped forward** — are rolled back simultaneously in a single backward-induction pass over $N = 500$ time steps (each price is the average of an $N$- and an $(N{+}1)$-step tree, cancelling the odd–even oscillation of the CRR lattice).

````python

def _bonus_cap_cert_price_one(S, K_cap, B_barrier, r, q, sigma, T, N):
    # Single CRR binomial tree pass (N steps).
    dt  = T / N
    u   = np.exp(sigma * np.sqrt(dt))
    d   = 1.0 / u
    p   = (np.exp((r - q) * dt) - d) / (u - d)
    df  = np.exp(-r * dt)

    j   = np.arange(N + 1)
    S_T = S * u ** (2 * j - N)

    V_do = np.maximum(K_cap - S_T, 0.0)
    V_do[S_T <= B_barrier] = 0.0

    V_cf = np.minimum(S_T, K_cap)

    for i in range(N - 1, -1, -1):
        V_do = df * (p * V_do[1:i + 2] + (1 - p) * V_do[0:i + 1])
        V_cf = df * (p * V_cf[1:i + 2] + (1 - p) * V_cf[0:i + 1])
        S_i  = S * u ** (2 * np.arange(i + 1) - i)
        V_do[S_i <= B_barrier] = 0.0

    return float(max(V_do[0] + V_cf[0], 0.0))


def bonus_cap_cert_price(S, K_cap, B_barrier, r, q, sigma, T, N=500):
    # Bonus Cap Certificate via binomial tree (CRR, N steps).
    # Cert = capped forward [min(S_T, K_cap)] + down-and-out put (K=K_cap, B=B_barrier).
    # Averages N and N+1 trees to cancel odd-even oscillation and smooth Greeks.
    if T <= 1e-8:
        return float(K_cap if S > B_barrier else min(S, K_cap))
    return 0.5 * (_bonus_cap_cert_price_one(S, K_cap, B_barrier, r, q, sigma, T, N) +
                  _bonus_cap_cert_price_one(S, K_cap, B_barrier, r, q, sigma, T, N + 1))
````

### Binomial Tree — Pricing Methodology

The certificate is priced via a **binomial tree with $N = 500$ steps** (each price averages the $N$- and $(N{+}1)$-step trees to smooth the CRR odd–even oscillation). The stock's evolution is discretised into $N$ equal time steps of length:

$$\Delta t = \frac{T}{N}$$

At each node the stock either moves **up** by factor $u$ or **down** by factor $d$, calibrated to match the stock's annualised volatility $\sigma$:

$$u = e^{\sigma\sqrt{\Delta t}}, \qquad d = \frac{1}{u} = e^{-\sigma\sqrt{\Delta t}}$$

Because $u \cdot d = 1$, the tree **recombines** — one up-move followed by one down-move lands at the same node as one down then one up. The final step has only $N+1$ distinct nodes instead of $2^N$.

#### Risk-Neutral Probability

Under a continuous dividend yield $\theta$, the no-arbitrage condition (lecture slide 59) is:

$$S_0 = e^{-r\,\Delta t}\left[q\cdot S_0\,u\,e^{\theta\,\Delta t} + (1-q)\cdot S_0\,d\,e^{\theta\,\Delta t}\right]$$

Solving for $q$ gives the **risk-neutral up-probability**:

$$q = \frac{e^{(r-\theta)\,\Delta t} - d}{u - d}$$

#### Tree Structure ($N = 3$ illustrative steps)

At step $n$ with $j$ up-moves ($0 \le j \le n$), the ex-dividend stock price at node $(n, j)$ is:

$$S_{n,j} = S_0 \cdot u^{\,j} \cdot d^{\,n-j}$$

The recombining lattice for $N = 3$:

| Step $n$ | $j=0$ | $j=1$ | $j=2$ | $j=3$ |
|:--------:|:-----:|:-----:|:-----:|:-----:|
| $n=0$ | $S_0$ | | | |
| $n=1$ | $S_0 d$ | $S_0 u$ | | |
| $n=2$ | $S_0 d^2$ | $S_0\underbrace{ud}_{=1}=S_0$ | $S_0 u^2$ | |
| $n=3$ | $S_0 d^3$ | $S_0 d^2 u$ | $S_0 d u^2$ | $S_0 u^3$ |

The middle node at $n=2$ equals $S_0 \cdot u \cdot d = S_0$ — the tree recombines back to the origin since $ud = 1$.

#### Backward Induction (three steps)

**1. Terminal payoffs** (at step $N$): assign each terminal node its derivative payoff $V_{N,j}$.

**2. Barrier knock-out**: at *every* step, nodes where $S_{n,j} \le B$ are absorbed — the down-and-out put value is forced to zero:

$$V_{n,j}^{\text{DO-put}} = 0 \qquad \text{if } S_{n,j} \le B$$

**3. Roll back one step**: for all surviving nodes, discount the risk-neutral expectation:

$$V_{n,j} = e^{-r\,\Delta t}\bigl[q\cdot V_{n+1,\,j+1} + (1-q)\cdot V_{n+1,\,j}\bigr]$$

Repeat steps 2–3 from $n = N-1$ down to $n = 0$. The root value $V_{0,0}$ is the fair certificate price.

### Task V – Daily Valuation Loop

Loads actual DU2076 market closing prices from `cert_prices.csv` (onvista.de export) and aligns them with yfinance trading dates over the **Sep 2025 – Jun 2026** window — 183 trading days, comfortably above the assignment's 100-trading-day minimum for the valuation window. For each trading day: (1) the remaining time to DU2076's maturity (18 Jun 2027) is computed; (2) the daily ECB Svensson spot rate evaluated at that day's remaining maturity `r(T_rem)` is looked up from `data/ECB Data Portal_20260615151519.csv`; (3) the 30-day rolling historical volatility `rolling_vol_30.asof(date)` is looked up for that day; and (4) the Bonus Cap Certificate is priced via the binomial tree ($N=500$, averaged with $N{+}1$). The resulting model-price series is stored alongside observed market prices for error analysis in the next cell.


````python
# ── Load actual DU2076 certificate prices (onvista.de export, Sep 2025 – Jun 2026) ─
cert_df = pd.read_csv(
    'data/cert_prices.csv', sep=';', encoding='utf-8-sig',
    decimal=',', thousands='.', dtype={'Datum': str}
)
cert_df['Datum'] = pd.to_datetime(cert_df['Datum'], format='%d.%m.%Y')
cert_df = cert_df.rename(columns={'Datum': 'Date', 'Schluss': 'Close'})
cert_df = cert_df.set_index('Date').sort_index()
cert_series = cert_df['Close'].dropna()

# ── Align cert dates to yfinance stock dates ──────────────────────────────────
common_dates     = prices.index.intersection(cert_series.index)
valuation_window = prices.loc[common_dates].copy()
cert_window      = cert_series.loc[common_dates].copy()

S_series  = valuation_window.values.flatten()
dates     = valuation_window.index
N         = len(S_series)
S_initial = float(S_series[0])

# ── DU2076 fixed parameters ───────────────────────────────────────────────────
K_cert = K_CERT    # from global params (DZ BANK KID)
B_cert = B_CERT    # from global params (DZ BANK KID)

print(f'Window : {dates[0].date()} to {dates[-1].date()}  ({N} days)')
print(f'S_initial = {S_initial:.2f},  K_cap = {K_cert:.2f},  B = {B_cert:.2f}')
print(f'Risk-free rate: daily ECB Svensson r(T_rem), maturity-matched per day')
print(f'Cert prices loaded: {len(cert_window)} days, {cert_window.iloc[0]:.2f} – {cert_window.iloc[-1]:.2f} EUR')

# ── Model prices: 30-day rolling vol, daily maturity-matched Svensson r ──────────────
model_prices = []
sigmas_used  = []
rates_used   = []

for i, (date, S) in enumerate(zip(dates, S_series)):
    T_rem     = max((MATURITY_DATE - date).days / 365, 1 / 365)
    sigma_est = float(np.clip(rolling_vol_30.asof(date), 0.10, 0.60))
    r_today   = get_risk_free_rate(date, T_rem)
    mp        = bonus_cap_cert_price(float(S), K_cert, B_cert, r_today, DIV_YIELD, sigma_est, T_rem)
    model_prices.append(mp)
    sigmas_used.append(sigma_est)
    rates_used.append(r_today)

model_prices = np.array(model_prices)
sigmas_used  = np.array(sigmas_used)
rates_used   = np.array(rates_used)

# ── Market prices: actual DU2076 closing prices from onvista.de ───────────────
market_prices = cert_window.values.astype(float)

print(f'30-day rolling vol: range {sigmas_used.min():.2%} – {sigmas_used.max():.2%}, mean {sigmas_used.mean():.2%}')
print(f'Mean Svensson r(T_rem): {rates_used.mean():.4%}  (range {rates_used.min():.4%} – {rates_used.max():.4%})')
print('Model and actual market prices computed.')
````

### Task V – Implied Volatility (Model-Implied vs. Realized)

Per the handwritten session notes on Task V ("implied VIX → implied is better, do both — realized is just nice to do"), this backs out the volatility that makes the binomial-tree price match the **observed DU2076 market price exactly** on each day — the derivatives-market analogue of Black-Scholes implied vol, applied to the certificate itself since no listed RHM option chain is available as a direct source. This is compared against the 30-day rolling **realized** volatility already used as the Task V pricing input, testing the "Task V – Results & Interpretation" claim below that realized vol overestimates what the market prices in.


````python
# ── Task V extension: Model-Implied Volatility (back-solved per day) ─────────
# DU2076 has negative vega (Task VI), so price(sigma) is monotonically
# decreasing in sigma: price(SIG_LO) is the achievable maximum and price(SIG_HI)
# the achievable minimum certificate value. If the observed market price falls
# outside that bracket (e.g. issuer margin pushes it above any achievable
# model price), no implied vol reproduces it exactly and the day is left NaN.
# A coarser N=120 tree is used here purely for solve speed (183 days x ~15-25
# brentq evaluations each); implied vol is a diagnostic/comparison, not the
# certificate's official Task V model price (which uses N=500 in the loop above).
from scipy.optimize import brentq

N_IMPLIED      = 120
SIG_LO, SIG_HI = 0.02, 2.00

def _implied_vol_for_day(S, mkt_price, r, q, T, K=K_cert, B=B_cert):
    f = lambda sig: bonus_cap_cert_price(S, K, B, r, q, sig, T, N=N_IMPLIED) - mkt_price
    lo_val, hi_val = f(SIG_LO), f(SIG_HI)
    if lo_val * hi_val > 0:
        return np.nan
    return brentq(f, SIG_LO, SIG_HI, xtol=1e-4, maxiter=60)

T_rem_all = np.array([max((MATURITY_DATE - d).days / 365, 1 / 365) for d in dates])
implied_vols = np.array([
    _implied_vol_for_day(float(S), float(mkt), r, DIV_YIELD, T)
    for S, mkt, r, T in zip(S_series, market_prices, rates_used, T_rem_all)
])

n_solved = int(np.sum(~np.isnan(implied_vols)))
vol_gap  = implied_vols - sigmas_used

print(f'Implied vol solved for {n_solved} / {len(implied_vols)} days '
      f'({len(implied_vols) - n_solved} days had no bracket -- market price outside the '
      f'achievable [sigma={SIG_HI:.0%}, sigma={SIG_LO:.0%}] price range, likely issuer margin)')
print(f'Mean implied vol               : {np.nanmean(implied_vols):.2%}')
print(f'Mean realized vol (30d rolling): {sigmas_used.mean():.2%}  (the Task V pricing input)')
print(f'Mean gap (implied - realized)  : {np.nanmean(vol_gap):+.2%}')

# ── Implied vs realized volatility chart (mirrors the lecture's DAX example) ─
fig, axes = plt.subplots(2, 1, figsize=(13, 7), sharex=True)

axes[0].plot(dates, sigmas_used * 100, color='steelblue', lw=1.5, label='Realized (30-day rolling)')
axes[0].plot(dates, implied_vols * 100, color='tomato', lw=1.2, alpha=0.85,
             label='Implied (back-solved from DU2076 market price)')
axes[0].set_ylabel('Volatility (% p.a.)')
axes[0].set_title('Task V — Implied vs. Realized Volatility (DU2076)', fontweight='bold')
axes[0].legend(fontsize=9)

axes[1].bar(dates, vol_gap * 100, color=['seagreen' if g >= 0 else 'tomato' for g in vol_gap],
            alpha=0.7, width=1)
axes[1].axhline(0, color='black', lw=0.5)
axes[1].axhline(np.nanmean(vol_gap) * 100, color='black', lw=1, linestyle='--',
                label=f'Mean gap = {np.nanmean(vol_gap)*100:+.1f} pp')
axes[1].set_ylabel('Implied − Realized (pp)')
axes[1].set_xlabel('Date')
axes[1].legend(fontsize=9)

plt.tight_layout()
plt.savefig('graphs/implied_vs_realized_vol.png', bbox_inches='tight')
plt.show()
````

## Task VI – Sensitivity Analysis / Greeks

Greeks are estimated numerically via central finite differences applied to `bonus_cap_cert_price`. A 1% spot bump is used for Delta and a 5% spot bump for Gamma — adjacent same-date lattice nodes at N = 500 lie ≈ 2σ√Δt ≈ 5.4% apart (≈ EUR 91 near the reference price), so a 1% bump is sub-node and the second difference collapses to float noise, while the 5% bump spans ≈ one node interval and, combined with the N/N+1 averaging and mild Gaussian smoothing, yields a stable Gamma. Vega uses a 2-percentage-point vol bump (binomial-tree noise makes smaller vol perturbations unreliable). Reference parameters: T = 1.79 y, σ = 44.7% (median 30-day rolling historical vol), S_ref = 1,708 EUR.

````python
# ── Numerical Greeks via central finite differences ───────────────────────────────────────────
# dS_gamma uses 5% because adjacent same-date lattice nodes at S≈1708 lie ~91 EUR apart
# (2*sigma*sqrt(dt), N=500); a 1% bump (~17 EUR) is sub-node and the second difference
# collapses to float noise, while the 5% bump (~85 EUR) spans ~one node interval.
dS    = 0.01   # 1% bump for delta
dSg   = 0.05   # 5% bump for gamma (must exceed ~one binomial step)
dsig  = 0.02   # 2% vol bump

def compute_greeks(S, K, B_barrier, r, q, sigma, T):
    f     = lambda s, sig, t: bonus_cap_cert_price(s, K, B_barrier, r, q, sig, t)
    price = f(S, sigma, T)
    delta = (f(S*(1+dS),  sigma, T) - f(S*(1-dS),  sigma, T)) / (2*dS*S)
    gamma = (f(S*(1+dSg), sigma, T) - 2*price + f(S*(1-dSg), sigma, T)) / (dSg*S)**2
    vega  = (f(S, sigma+dsig, T) - f(S, sigma-dsig, T)) / (2*dsig) / 100
    return {'price': price, 'delta': delta, 'gamma': gamma, 'vega': vega}

# Reference parameters: first DU2076 trading day (04 Sep 2025)
T_ref   = round(max((MATURITY_DATE - dates[0]).days / 365, 1/365), 2)
sig_ref = float(np.median(sigmas_used))
S_ref   = S_initial

S_vals    = np.linspace(B_cert * 0.55, K_cert * 1.55, 200)
results_g = [compute_greeks(s, K_cert, B_cert, RISK_FREE, DIV_YIELD, sig_ref, T_ref)
             for s in S_vals]

from scipy.ndimage import gaussian_filter1d
deltas = gaussian_filter1d(np.array([g['delta'] for g in results_g]), sigma=2)
gammas = gaussian_filter1d(np.array([g['gamma'] for g in results_g]), sigma=2)
vegas  = gaussian_filter1d(np.array([g['vega']  for g in results_g]), sigma=2)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle(
    f'Task VI – Greeks of DU2076 Bonus Cap Certificate  '
    f'(T={T_ref:.2f}y, σ={sig_ref:.1%}, B={B_cert:.0f}, K={K_cert:.0f})',
    fontweight='bold', fontsize=12)

plot_data = [
    (deltas, 'Delta  dV/dS',             'steelblue'),
    (gammas, 'Gamma  d²V/dS²', 'seagreen'),
    (vegas,  'Vega  dV/dσ (per 1% vol)', 'darkorange'),
]

for ax, (vals, name, color) in zip(axes, plot_data):
    ax.plot(S_vals, vals, color=color, lw=2)
    ax.axvline(B_cert, color='red',   lw=1, linestyle='--', alpha=0.7, label=f'B={B_cert:.0f}')
    ax.axvline(K_cert, color='green', lw=1, linestyle='--', alpha=0.7, label=f'K={K_cert:.0f}')
    ax.axvline(S_ref,  color='gray',  lw=1, linestyle=':',  alpha=0.5, label=f'S_ref={S_ref:.0f}')
    ax.set_xlabel('Stock price S')
    ax.set_title(name, fontweight='bold')
    ax.legend(fontsize=8)
    if name.startswith('Gamma'):
        ax.set_ylim(-0.005, 0.008)
    elif name.startswith('Vega'):
        ax.set_ylim(-20, 2)

plt.tight_layout()
plt.savefig('graphs/greeks.png', bbox_inches='tight')
plt.show()

g_now = compute_greeks(S_ref, K_cert, B_cert, RISK_FREE, DIV_YIELD, sig_ref, T_ref)
print('Greeks at reference stock price:')
for k, v in g_now.items():
    print(f'  {k:8s} = {v:.6f}')
# Theta (365-day convention) and Rho at the reference point — reserved for the
# oral presentation (the poster shows Delta/Gamma/Vega per the brief).
_f_ref    = lambda T=T_ref, r=RISK_FREE: bonus_cap_cert_price(S_ref, K_cert, B_cert, r, DIV_YIELD, sig_ref, T)
h_T, h_r  = 5 / 365, 0.005
theta_day = -(_f_ref(T=T_ref + h_T) - _f_ref(T=T_ref - h_T)) / (2 * h_T) / 365
rho_1pp   =  (_f_ref(r=RISK_FREE + h_r) - _f_ref(r=RISK_FREE - h_r)) / (2 * h_r) / 100
print(f'  theta    = {theta_day:+.4f}  EUR per calendar day (365-day convention)')
print(f'  rho      = {rho_1pp:+.4f}  EUR per +1pp interest rate')
print('Negative Vega: higher vol raises barrier-breach probability, reducing the bonus payout.')
````

````python
# ── Theta and Rho across spot (oral-presentation supplement) ──────────────────
def theta_rho(S, K, B_barrier, r, q, sigma, T):
    theta = -(bonus_cap_cert_price(S, K, B_barrier, r, q, sigma, T + h_T)
              - bonus_cap_cert_price(S, K, B_barrier, r, q, sigma, T - h_T)) / (2 * h_T) / 365
    rho   =  (bonus_cap_cert_price(S, K, B_barrier, r + h_r, q, sigma, T)
              - bonus_cap_cert_price(S, K, B_barrier, r - h_r, q, sigma, T)) / (2 * h_r) / 100
    return theta, rho

tr_results = [theta_rho(s, K_cert, B_cert, RISK_FREE, DIV_YIELD, sig_ref, T_ref) for s in S_vals]
thetas = gaussian_filter1d(np.array([t for t, _ in tr_results]), sigma=2)
rhos   = gaussian_filter1d(np.array([r for _, r in tr_results]), sigma=2)

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
fig.suptitle(
    f'Task VI – Theta & Rho of DU2076 Bonus Cap Certificate  '
    f'(T={T_ref:.2f}y, σ={sig_ref:.1%}, B={B_cert:.0f}, K={K_cert:.0f})',
    fontweight='bold', fontsize=12)

plot_data = [
    (thetas, 'Theta  dV/dt (EUR per calendar day)', 'crimson'),
    (rhos,   'Rho  dV/dr (per +1pp rate)',           'indigo'),
]

for ax, (vals, name, color) in zip(axes, plot_data):
    ax.plot(S_vals, vals, color=color, lw=2)
    ax.axvline(B_cert, color='red',   lw=1, linestyle='--', alpha=0.7, label=f'B={B_cert:.0f}')
    ax.axvline(K_cert, color='green', lw=1, linestyle='--', alpha=0.7, label=f'K={K_cert:.0f}')
    ax.axvline(S_ref,  color='gray',  lw=1, linestyle=':',  alpha=0.5, label=f'S_ref={S_ref:.0f}')
    ax.set_xlabel('Stock price S')
    ax.set_title(name, fontweight='bold')
    ax.legend(fontsize=8)

plt.tight_layout()
plt.savefig('graphs/theta_rho.png', bbox_inches='tight')
plt.show()
````

**Gamma near reference:** Gamma is negative as S approaches the cap K = 2,000 — the certificate is implicitly short an OTM call, so Delta falls as S rises above the cap, opposite to a vanilla long. Near the barrier, Gamma spikes sharply as the knock-out probability rises steeply with proximity to B = 1,050, making the Delta surface highly convex there.

**Vega:** Negative Vega comes from two simultaneous sources. The capped forward structure makes the certificate effectively short a call at K = 2,000; higher vol raises that call’s value, hurting the short position. Simultaneously, higher vol raises the probability that Rheinmetall touches the barrier at B = 1,050, permanently eliminating the bonus. Both effects point the same way: the holder of this certificate wants low realised volatility.

**Theta and Rho:** Theta is *positive* (+0.36 EUR/day, 365-day convention) — each passing day shrinks the window in which the barrier can be hit and pulls the fixed €2,000 redemption closer, the opposite of a long vanilla option's time decay. Rho is *negative* (≈ −12.4 EUR per +1pp) — the certificate's value is dominated by the discounted bond-like bonus payment, so higher rates reduce present value. Per the brief, the poster shows Delta/Gamma/Vega; Theta/Rho are reserved for the oral presentation.


## Task VII – Replicating Portfolio

The replicating portfolio at each date $t$ holds $\Delta(t)$ shares of Rheinmetall and a bond position $B(t)$ — exactly as in the lecture (slide 29):

$$V(t) = \underbrace{\Delta(t)\cdot S(t)}_{\text{equity leg}} + \underbrace{B(t)}_{\text{bond leg}}, \qquad B(t) = V(t) - \Delta(t)\cdot S(t)$$

$\Delta(t)$ is obtained from the binomial-tree price by a central finite difference (±1% spot bump via `compute_greeks`) — the numerical equivalent of the lecture's one-period branch ratio (slide 29):

$$\Delta = \frac{C_u - C_d}{S_u - S_d}$$

The **equity fraction** measures what share of the certificate's total value the stock leg represents on a given day:

$$\text{Equity Fraction}(t) = \frac{\Delta(t)\cdot S(t)}{V(t)}$$

This fraction can exceed 100% near the barrier — when Delta rises toward 1.0 and the certificate value is falling, the stock leg as a fraction of value overshoots. The bond component turns negative (short bond, long more than one unit of stock), which is the replicating portfolio's way of capturing the heightened stock sensitivity near B. Both $\Delta(t)$ and $V(t)$ use the same day-by-day 30-day rolling historical volatility as the Task V pricing loop, together with the constant ECB Svensson benchmark rate.


````python
# ── Daily replicating portfolio: Delta shares + Bond ──────────────────────────
rep_deltas  = []
rep_bonds   = []
equity_frac = []

for i, (date, S) in enumerate(zip(dates, S_series)):
    T_rem     = max((MATURITY_DATE - date).days / 365, 1 / 365)
    sigma_est = float(np.clip(sigmas_used[i], 0.10, 0.60))

    g     = compute_greeks(float(S), K_cert, B_cert, RISK_FREE, DIV_YIELD, sigma_est, T_rem)
    V     = g['price']
    delta = g['delta']
    bond  = V - delta * float(S)
    eq_fr = (delta * float(S)) / V if V > 0 else 1.0

    rep_deltas.append(delta)
    rep_bonds.append(bond)
    equity_frac.append(eq_fr)

rep_deltas  = np.array(rep_deltas)
rep_bonds   = np.array(rep_bonds)
equity_frac = np.array(equity_frac)
````

### Task VII – Time Series: Delta and Equity Fraction

The chart above (`replicating_portfolio_time.png`) shows how both Delta and the equity fraction evolved across the 183-day window as Rheinmetall fell from EUR 1,708 in early September 2025 toward the barrier region in spring 2026.

**Delta panel:** Delta started around 0.60 on the first trading day (S ≈ 1,708, 30-day vol ≈ 43%) and initially *fell* to its window low of ≈ 0.14 as Rheinmetall rallied toward €1,900–1,950 in September–October 2025 — with the stock approaching the cap and vol easing, the certificate increasingly resembled its fixed bonus leg and lost directional sensitivity. As the stock declined through late 2025 and into 2026, Delta climbed steadily. In the barrier-approach period around May 2026 it reached ≈ 0.9–1.1: the certificate was essentially a 1:1 (at times slightly leveraged) long position in Rheinmetall, with almost no residual protection. An investor who thought they were "covered" would have been surprised by this near-total stock exposure at the worst point of the drawdown.

**Equity fraction panel:** The equity fraction (Δ·S / V) started near 70%, fell to its low of ≈ 16% during the autumn 2025 rally, and then climbed persistently as the stock declined — it exceeded 50% on 143 of the 183 days. In the May 2026 barrier-approach period it rose above 100% (peak ≈ 112%): the bond leg turned *negative*, i.e. the replicating portfolio was short the bond and held more than the certificate's full value in stock. A continued decline through 1,050 would have triggered the knock-out and converted the position into a capped forward.

**Key falling-market point:** Unlike a capital-protected note — where Delta tends toward 0 as S falls, reducing the investor's stock exposure — a Bonus Certificate's Delta rises toward 1.0 as the barrier approaches. That is the structural difference between the two product types: the barrier creates a conditional exposure that amplifies stock sensitivity at exactly the worst time. Anyone holding DU2076 in a sustained decline has a decision to make well before the barrier is reached, because once it is breached the protection is gone permanently and the certificate converts to a capped forward with full downside and no floor.


### Task VII – Equity Fraction vs. Stock Price (Static Snapshot)

The plot below sweeps stock price from 0.5·B to 1.6·K at a fixed T_rem ≈ 1.79 y (first DU2076 trading day, 04 Sep 2025; [DZ BANK KID](https://bib-service.dzbank.de/bib/DE000DU20767-de-DE.pdf)) and the median 30-day rolling historical vol over the valuation window (`sig_snap = np.median(sigmas_used)`). The purpose is to show the equity fraction purely as a function of the stock level, independent of the time dimension already captured above.

Above the barrier the binomial-tree Delta applies: the fraction peaks sharply near B = 1,050 (the barrier-approach zone where Delta → 1 and the bond leg turns negative) and then declines as the stock moves comfortably into the bonus zone. At very high stock prices, well above the bonus level, the fraction falls toward zero as the certificate behaves increasingly like the bond component of its structure. Below the barrier the certificate has knocked out and reverts to a capped forward, $S e^{-qT} - \text{Call}(K{=}2{,}000)$, whose delta is $e^{-qT} - \Delta_{\text{call}}$ — the still-alive short cap call dampens roughly a quarter of the stock exposure just below the barrier. The equity fraction therefore *drops* discontinuously at the knock-out (from > 100% on the live certificate to ≈ 81%) and then climbs back toward ≈ 98% as S falls further and the short call fades: the holder bears near-full downside with no remaining protection, moderated only by the fading short-call premium.


````python
# ── Equity fraction vs S (static snapshot at T_rem = T_ref) ─────────────────
# T_snap reuses T_ref: remaining maturity on the first DU2076 trading day (04 Sep 2025)
# Source: DZ BANK KID — maturity date 18 Jun 2027 (https://www.dzbank-wertpapiere.de/DU2076)
T_snap   = T_ref
sig_snap = float(np.median(sigmas_used))

S_cross  = np.linspace(B_cert * 0.50, K_cert * 1.60, 300)  # brackets barrier (B=1,050) and cap (K=2,000) with margin
eq_cross = []

for s in S_cross:
    if s <= B_cert:
        # knocked-out: reverts to capped forward = S*exp(-qT) - Call(K_cap)
        # equity fraction = (delta_capped_fwd * s) / fwd_val
        # delta of capped forward = exp(-qT) - BS call delta (short cap call still alive)
        fwd_val  = s * np.exp(-DIV_YIELD * T_snap) - bs_vanilla_call(s, K_cert, RISK_FREE, DIV_YIELD, sig_snap, T_snap)
        fwd_val  = max(fwd_val, 1e-6)
        _d1      = (np.log(s / K_cert) + (RISK_FREE - DIV_YIELD + 0.5 * sig_snap**2) * T_snap) / (sig_snap * np.sqrt(T_snap))
        delta_ko = np.exp(-DIV_YIELD * T_snap) * (1.0 - norm.cdf(_d1))
        eq_cross.append((delta_ko * s) / fwd_val)
        continue
    g = compute_greeks(s, K_cert, B_cert, RISK_FREE, DIV_YIELD, sig_snap, T_snap)
    V = g['price']
    d = g['delta']
    eq_cross.append((d * s / V) if V > 0 else 1.0)

eq_cross = np.array(eq_cross)

# Smooth the tree-induced sawtooth ONLY in the live region (S > B_cert).
# The binomial delta jumps as S sweeps across discrete tree nodes; a 1-D Gaussian
# filter (sigma=2, matching Task VI) removes that numerical sawtooth.
# NB: the knocked-out branch (S <= B_cert) is a genuine economic discontinuity and
# is left untouched so the barrier jump at S = B_cert is not blurred away.
from scipy.ndimage import gaussian_filter1d
live_mask = S_cross > B_cert
eq_cross[live_mask] = gaussian_filter1d(eq_cross[live_mask], sigma=2)

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(S_cross, eq_cross * 100, color='steelblue', lw=2)
ax.axvline(B_cert, color='red',   lw=1.5, linestyle='--', label=f'Barrier B={B_cert:.0f}')
ax.axvline(K_cert, color='green', lw=1.5, linestyle='--', label=f'Bonus K={K_cert:.0f}')
ax.set_xlabel('Stock price S')
ax.set_ylabel('Equity fraction (%)')
ax.set_title(f'Task VII – Equity Fraction vs S  (T_rem={T_snap}y, sigma={sig_snap:.1%})',
             fontweight='bold')
ax.legend()
ax.set_ylim(0, 135)

plt.tight_layout()
plt.savefig('graphs/equity_fraction_vs_S.png', bbox_inches='tight')
plt.show()

print('Interpretation:')
print('  As S falls toward the barrier, equity fraction rises above 100% (leveraged).')
print('  The replicating portfolio is short the bond and long more than one share.')
print('  This reflects the strongly positive delta of the down-and-out put near the')
print('  barrier: a small up-move away from B restores the entire bonus claim.')
print('  In a falling market, the certificate behaves like a leveraged stock position')
print('  until the barrier is breached, at which point protection is permanently lost.')
print('  Below the barrier, the certificate has already knocked out and reverts to a')
print('  capped forward: the equity fraction drops to ~81% just under B (the short cap')
print('  call still dampens ~25% of the exposure) and climbs back toward ~98% as S')
print('  falls further — near-full downside with no remaining protection.')
````

### Monte Carlo Simulation – Risk-Neutral Path Generation

Simulates 50,000 independent one-year trajectories of Rheinmetall AG's stock price at daily frequency (252 steps), evolved under the **risk-neutral measure**: drift = r − q − ½σ², per the course's Monte Carlo recipe (Slide 145 / "Simulating paths" slide). σ is the 252-day trailing historical volatility estimated from the yfinance log-return series above; the historical mean return μ = 54.05% p.a. is reported for context only and does not enter the simulation. The random seed is fixed at 2026 for full reproducibility. All four subsequent tasks (VIII, IX, X, XI) reuse this single shared path set so that any differences in outcomes across strategies are purely due to the strategy, not to Monte Carlo noise.


````python
# ── Monte Carlo parameters ────────────────────────────────────────────────────
W0       = 10_000
T_star   = 1.0
N_STEPS  = 252
N_PATHS  = 50_000
S0_mc    = float(prices.iloc[-1])
mu_hist  = float(log_returns.mean() * 252)   # reference only — NOT the simulation drift (see note below)
sigma_mc = float(log_returns.loc[:'2026-06-01'].iloc[-252:].std() * np.sqrt(252))  # 252-day trailing historical vol to 01 Jun 2026

# ── Risk-neutral drift, per the course's MC recipe ("Drift parameter -> risk
# neutral measure!" / "Simulating paths" slide): d ln S = (r - q - 1/2 sigma^2) dt + z sigma sqrt(dt)
mu_rn = RISK_FREE - DIV_YIELD - 0.5 * sigma_mc**2

print('Monte Carlo Setup:')
print(f'  S0            = EUR {S0_mc:.2f}')
print(f'  r  (drift)    = {RISK_FREE:.2%}   [risk-neutral rate; drives the simulation drift]')
print(f'  q  (dividend) = {DIV_YIELD:.2%}')
print(f'  sigma         = {sigma_mc:.2%}   (252-day trailing historical)')
print(f'  mu hist       = {mu_hist:.2%}   [reference only, NOT used as the simulation drift]')
print(f'  Paths         = {N_PATHS:,}')

# ── Simulate paths (risk-neutral log-price recipe, Slide 145) ────────────────
dt_mc = T_star / N_STEPS
np.random.seed(2026)

Z          = np.random.standard_normal((N_PATHS, N_STEPS))
increments = mu_rn * dt_mc + sigma_mc * np.sqrt(dt_mc) * Z
log_paths  = np.cumsum(increments, axis=1)
S_paths    = S0_mc * np.exp(np.hstack([np.zeros((N_PATHS, 1)), log_paths]))
S_terminal = S_paths[:, -1]

print(f'Simulation complete.')
print(f'Terminal price range: {S_terminal.min():.1f} – {S_terminal.max():.1f}')
````

### Modelling Conventions for Tasks VIII–XI

The paths generated above evolve the log-price under the **risk-neutral measure**, exactly as specified for Part 3 in the session notes ("Drift parameter → risk neutral measure!" / "Simulating paths": Δ ln S = (r − q − ½σ²)·Δt + σ·√Δt·Z), using the same `RISK_FREE` (r) and `DIV_YIELD` (q) that price the puts via Black–Scholes elsewhere in the notebook. This keeps the underlying simulation and the option pricing consistent under a single measure, and avoids relying on Rheinmetall's realised 54.05% p.a. historical mean return, which reflects an idiosyncratic 2023–2026 re-rating rather than a defensible forward-looking estimate (`mu_hist` is retained above for context only).

**Why the baseline looks so risky.** At σ ≈ 45.6%, the volatility-drag term −½σ² ≈ −10.4% p.a. dominates the risk-free rate r − q ≈ 2.1%, so the risk-neutral **log**-drift of the simulation is **−8.27% p.a.** — meaningfully negative, even though nothing in the real world is assumed to be falling. This is a standard feature of risk-neutral (Q-measure) simulation, not a real-world forecast for Rheinmetall: Q-measure paths are constructed so the *discounted* price is a martingale, not to be an unbiased predictor of the *actual* future price. The practical consequence for Task VIII is that the unhedged position still shows a large 95% VaR (**56.38%**) purely from this high-volatility construction — even though, once returns are stated in simple (arithmetic) terms (see the return convention below), the *mean* return comes out a small **positive +2.16%**, matching the closed-form e^{r−q} − 1 ≈ 2.14% almost exactly, as an arbitrage-free risk-neutral asset should. It is the large VaR — not a negative mean — that makes the put overlay in Tasks IX–XI valuable: the terminal-wealth distribution is heavily right-skewed with a long left tail even though its average outcome is unbiased.

Dividends enter both the simulation drift and the option pricing consistently: `bs_vanilla_put` takes the dividend yield q, and the underlying path drift includes the same −q term.

**Return convention:** all Part 3 performance and risk figures (mean, std, skewness, VaR, CVaR) are stated in **simple (arithmetic) wealth return**, `wealth_T / wealth_0 − 1`, not continuously-compounded log return. The underlying simulation itself still evolves in log-increments under the risk-neutral measure exactly as described above (that mechanics is unchanged) — but before any statistic is computed, the reporting layer converts terminal wealth to a simple return. The Task XI 15% ceiling is therefore a direct 15% cap on simple/relative wealth loss — no log-to-simple translation factor is needed, since VaR is already expressed as a wealth-loss fraction.


## Task VIII – Performance Analysis (No Risk Management)

Defines the reusable `performance_stats` helper, trimmed to five headline performance/risk measures — mean, standard deviation, skewness, 95% VaR, and 95% CVaR — per the assignment's "4-5 measures for a good grade" guidance (median, excess kurtosis, Sharpe and P(loss) are dropped; kurtosis is explicitly flagged as optional in the same guidance). It is then applied first to the fully unhedged stock position (100% of W₀ = €10,000 in Rheinmetall), which serves as the baseline against which all insurance strategies in Tasks IX–XI are compared.

````python
from scipy.stats import skew as _skew

def performance_stats(returns, label=''):
    # Print and return the 5 headline performance/risk measures.
    mean_r   = np.mean(returns)
    std_r    = np.std(returns)
    skewness = _skew(returns)
    var95    = np.percentile(returns, 5)          # 5th pct = 95% VaR
    cvar95   = np.mean(returns[returns <= var95])

    if label:
        print(f'--- {label} ---')
    print(f'  Mean return       : {mean_r:>8.2%}')
    print(f'  Std deviation     : {std_r:>8.2%}')
    print(f'  Skewness          : {skewness:>8.3f}')
    print(f'  95% VaR (loss)    : {-var95:>8.2%}')
    print(f'  95% CVaR (loss)   : {-cvar95:>8.2%}')
    print()
    return {'mean': mean_r, 'std': std_r, 'skewness': skewness,
            'var95': var95, 'cvar95': cvar95}


returns_unhedged = S_terminal / S0_mc - 1     # simple (arithmetic) wealth return, not log return
stats_uh = performance_stats(returns_unhedged, 'Task VIII – Unhedged (pure stock)')
print(f'MC precision: {N_PATHS:,} paths; standard error of the mean return = '
      f'{returns_unhedged.std() / np.sqrt(N_PATHS):.3%}')

expected_mean = np.exp(RISK_FREE - DIV_YIELD) - 1
print(f'Closed-form check: E[S_T/S0]-1 = e^(r-q)-1 = {expected_mean:.4%}  '
      f'vs. simulated mean = {stats_uh["mean"]:.4%}')
assert abs(stats_uh['mean'] - expected_mean) < 5 * (returns_unhedged.std() / np.sqrt(len(returns_unhedged))), \
    'Unhedged mean simple return should match e^(r-q)T - 1 within MC noise'

fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(returns_unhedged * 100, bins=200, density=True,
        color='steelblue', alpha=0.7, label='Unhedged')
ax.axvline(stats_uh['var95'] * 100, color='red', lw=2, linestyle='--',
           label=f"95% VaR = {-stats_uh['var95']:.1%}")
ax.axvline(0, color='black', lw=1, alpha=0.5)
ax.set_xlabel('1-Year Return (%)')
ax.set_ylabel('Density')
ax.set_title('Task VIII – Return Distribution (No Hedging)', fontweight='bold')
ax.legend()
plt.tight_layout()
plt.savefig('graphs/mc_unhedged.png', bbox_inches='tight')
plt.show()
````

## Task IX – Portfolio Insurance with Put Options

Defines `insured_return`, which allocates fraction α of W₀ = €10,000 to European put options (priced at issuance via Black-Scholes on historical vol — the "fictitious puts" specification in the assignment) and the remainder to the underlying stock. Terminal wealth is computed from the 50,000 simulated terminal prices. The function is evaluated across a 4×4 grid of α ∈ {5%, 10%, 15%, 20%} and K ∈ {90%, 95%, 100%, 105%} of S₀, yielding 16 risk-return combinations for comparison in the heatmaps below. The grid spans the practically relevant range for retail portfolio insurance — 5–20% allocation budgets and 90–105% strikes — consistent with the configurations studied in the foundational portfolio insurance literature ([Rubinstein & Leland, *J. Finance* 1981](https://doi.org/10.1111/j.1540-6261.1981.tb00673.x); [Leland, *J. Finance* 1985](https://doi.org/10.1111/j.1540-6261.1985.tb02382.x)).

````python
def insured_return(S_term, S0, alpha, K_put, r, q, sigma_price, T, W=10_000):
    # Return distribution (simple/arithmetic wealth return) for (1-alpha)*stock + alpha*puts portfolio.
    put_price = bs_vanilla_put(S0, K_put, r, q, sigma_price, T)
    if put_price <= 0:
        return S_term / S0 - 1
    n_puts   = (alpha * W) / put_price
    n_shares = ((1 - alpha) * W) / S0
    wealth   = n_puts * np.maximum(K_put - S_term, 0) + n_shares * S_term
    return wealth / W - 1


# Grid: 5–20% allocation (typical retail insurance budget) × 90–105% strikes
# (OTM to slightly ITM); range consistent with portfolio insurance literature
# (Rubinstein & Leland, J. Finance 1981; Leland, J. Finance 1985)
alpha_vals  = [0.05, 0.10, 0.15, 0.20]
strike_pcts = [0.90, 0.95, 1.00, 1.05]

results_ix = {}
rows = []

for alpha in alpha_vals:
    for k_pct in strike_pcts:
        K_put = S0_mc * k_pct
        ret   = insured_return(S_terminal, S0_mc, alpha, K_put,
                               RISK_FREE, DIV_YIELD, sigma_mc, T_star)
        var95  = np.percentile(ret, 5)
        cvar95 = np.mean(ret[ret <= var95])
        std_r  = np.std(ret)

        results_ix[(alpha, k_pct)] = ret
        rows.append({
            'Alpha': f'{alpha:.0%}',
            'Strike': f'{k_pct:.0%}',
            'Mean': f"{np.mean(ret):.2%}",
            'Std': f"{std_r:.2%}",
            'Skew': f'{_skew(ret):.2f}',
            '95% VaR': f"{-var95:.2%}",
            '95% CVaR': f"{-cvar95:.2%}",
        })

print('Task IX – Insurance Strategy Summary')
print(pd.DataFrame(rows).to_string(index=False))
````

### Task IX – Discussion of Observations

Under the risk-neutral simulation (log-drift ≈ **−8.27%** p.a. at σ ≈ 45.6%, see *Modelling Conventions* above), reported in simple (arithmetic) wealth return, the **unhedged** position carries a 95% VaR of **56.38%** — still a large baseline driven by the volatility-drag term, though far below the raw log-return VaR because simple return is bounded below at −100% while log return is not. Against this baseline, even a modest put sleeve delivers a substantial risk reduction: at α = 5% VaR sits around 38–43% depending on strike; by α = 15% it reaches the grid minimum of **14.73%** (K = 100%), roughly a 74% reduction in tail risk from the unhedged position.

**Reading down allocation rows (α):** VaR falls steeply from α = 5% to α = 10% at every strike (e.g. K = 100%: 41.63% → 26.88%). Past that point the pattern **depends on strike**. At the cheaper OTM strikes (90%, 95%) VaR is U-shaped — it troughs around α = 10–15% and drifts back up by α = 20% (e.g. K = 90%: 20.54% at α = 10% → 25.83% at α = 20%), because the fixed premium budget buys progressively fewer contracts as α grows relative to a put with a small payoff per contract. K = 100% shows the same U-shape but with its trough at α = 15% (14.73%) before ticking back up to 18.46% at α = 20%. Only the priciest strike, K = 105%, keeps falling all the way to α = 20% (29.30% → 15.77% → 14.92%), because each additional euro of premium there buys a contract with a materially larger payoff at the 5th-percentile outcome.

**Reading across strikes (K):** the ranking flips with α. At low allocations (5%, 10%) the cheapest, deepest-OTM strike (90%) gives the *lowest* VaR (e.g. α = 10%: K = 90% → 20.54% vs. K = 105% → 29.30%) — more contracts, more aggregate tail payoff, for the same budget. By α = 15% the near-the-money K = 100% strike takes over as best (14.73%, narrowly ahead of K = 105% at 15.77%); by α = 20% the ranking flips once more and K = 105% becomes best (14.92% vs. 18.46% for K = 100%), since the investor can now afford enough of the more expensive, more protective contracts. The single lowest-VaR cell in the grid is **α = 15%, K = 100%** (VaR = 14.73%), narrowly ahead of α = 20%, K = 105% (14.92%) — both sit right at the edge of the Task XI 15% ceiling.

**Skewness:** unlike log return, simple-return skewness is already substantial at the **unhedged baseline (1.60)** — a lognormal terminal-price distribution is inherently right-skewed in level/simple-return space even though it is close to symmetric in log space. Insurance pushes it materially higher still, generally rising with both α and K, from roughly 2.0–2.1 at α = 5% up to 3.0–3.6 at α = 15–20% (K ≥ 95%) — the put floor truncates the left tail on top of an already right-skewed base distribution, rather than starting from near-zero skew as a log-return view would suggest.

**Mean return:** every cell in the grid — including the unhedged baseline (+2.16%) — now has a small **positive** mean simple return, rising mildly with α (from ≈2.20% at α = 5% to ≈2.33–2.34% at α = 20%) and essentially flat across strike within a given α row. This is because the put leg, correctly priced off (r, q, σ), carries expected payoff growth tied to r, while the stock leg's expected value grows at r − q; shifting the budget toward puts mechanically raises expected wealth by roughly q per unit reallocated. Mean return is therefore no longer degenerate, but VaR and CVaR remain the operative decision criteria here, consistent with the "meaningful measures" the session notes flag (risk and tail-risk measures), alongside skewness as the one descriptive moment retained.

**Read-through to Task XI:** because the unhedged VaR (56.38%) is still well above the 15% ceiling, a substantial put allocation is required to comply. Task XI confirms that only the near-the-money strikes (100%, 105%) can reach the ceiling within a practical allocation budget.


## Task X – Stress Scenario Analysis

Tests a representative mid-grid allocation (α = 10%, K = 95% of S₀) from the Task IX grid under six scenarios (the fifth and sixth being the combined "and" cases): (1) baseline simulated paths; (2) put pricing vol raised by +5pp; (3) put pricing vol lowered by −5pp; (4) a −20% price shock at the mid-horizon (T/2); (5)–(6) the shock combined with each vol direction. Scenarios 2–3 quantify the impact of volatility mis-estimation at put issuance; Scenario 4 tests resilience to a sudden Rheinmetall price crash before the options expire, and Scenarios 5–6 stack both shocks together (the jointly adverse and jointly offsetting "and" cases). For each scenario the five headline measures — mean, standard deviation, skewness, 95% VaR and 95% CVaR — are tabulated, and the scenario VaR lines are overlaid on the return-distribution chart.

````python
# ── Chosen allocation for stress test ────────────────────────────────────────
alpha_s = 0.10    # 10% in puts
k_pct_s = 0.95    # strike = 95% of S0
K_s     = S0_mc * k_pct_s

# Baseline
ret_base   = insured_return(S_terminal, S0_mc, alpha_s, K_s,
                             RISK_FREE, DIV_YIELD, sigma_mc, T_star)

# Stress 1: pricing vol +5pp — model-risk sensitivity for options;
# ±5pp is the standard vega stress magnitude under FRTB/Basel IV
# (BCBS, Minimum capital requirements for market risk, Jan 2019, §21.83)
ret_vol_up = insured_return(S_terminal, S0_mc, alpha_s, K_s,
                             RISK_FREE, DIV_YIELD, sigma_mc + 0.05, T_star)

# Stress 2: pricing vol -5pp (same FRTB magnitude, downward direction)
ret_vol_dn = insured_return(S_terminal, S0_mc, alpha_s, K_s,
                             RISK_FREE, DIV_YIELD, sigma_mc - 0.05, T_star)

# Stress 3: −20% price shock at T/2 — adverse scenario consistent with
# PRIIPs KID stress scenario methodology (EU Regulation 1286/2014;
# ESMA/2017/1253); −20% is a representative single-stock adverse return
# at 6-month horizon for high-vol equities (~48% annualised vol)
S_paths_shock = S_paths.copy()
mid           = N_STEPS // 2
S_paths_shock[:, mid:] *= 0.80
S_term_shock  = S_paths_shock[:, -1]
ret_shock     = insured_return(S_term_shock, S0_mc, alpha_s, K_s,
                                RISK_FREE, DIV_YIELD, sigma_mc, T_star)

# Stress 4 (combined "and" case): -20% price shock at T/2 AND pricing vol +5pp
# together — a crash coinciding with a volatility spike, the jointly adverse
# scenario the "and/or" wording asks us to consider.
ret_combined  = insured_return(S_term_shock, S0_mc, alpha_s, K_s,
                                RISK_FREE, DIV_YIELD, sigma_mc + 0.05, T_star)

# Stress 5 (combined "and" case, opposite vol direction): -20% price shock at
# T/2 AND pricing vol -5pp together. Completes the full stress-combination set
# from the session notes (+5%, -5%, +5%&-20%, -5%&-20%, -20% alone).
ret_combined_dn = insured_return(S_term_shock, S0_mc, alpha_s, K_s,
                                  RISK_FREE, DIV_YIELD, sigma_mc - 0.05, T_star)

scenarios = {
    'Baseline':              ret_base,
    'Vol +5pp':              ret_vol_up,
    'Vol -5pp':              ret_vol_dn,
    '-20% shock at T/2':     ret_shock,
    '-20% shock & vol +5pp': ret_combined,
    '-20% shock & vol -5pp': ret_combined_dn,
}

print('=' * 74)
print(f'  TASK X – Stress Analysis  (alpha={alpha_s:.0%}, K={k_pct_s:.0%} of S0)')
print('=' * 74)
print(f'{"Scenario":<25} {"Mean":>8} {"Std":>8} {"Skew":>6} {"VaR95":>8} {"CVaR95":>9}')
print('-' * 74)
for name, ret in scenarios.items():
    m    = np.mean(ret)
    sd   = np.std(ret)
    sk   = _skew(ret)
    v95  = np.percentile(ret, 5)
    cv95 = np.mean(ret[ret <= v95])
    print(f'{name:<25} {m:>7.2%} {sd:>7.2%} {sk:>6.2f} {-v95:>7.2%}  {-cv95:>8.2%}')

# ── Stress distribution chart with VaR vertical lines ────────────────────────
fig, ax = plt.subplots(figsize=(11, 6))
colors_sc = ['steelblue', 'darkorange', 'seagreen', 'tomato', 'purple', 'gray']
for (name, ret), color in zip(scenarios.items(), colors_sc):
    ax.hist(ret * 100, bins=150, density=True, alpha=0.45, color=color, label=name)
    v = np.percentile(ret, 5)
    ax.axvline(v * 100, color=color, lw=1.2, linestyle='--', alpha=0.8)
ax.axvline(0, color='black', lw=1, alpha=0.5)
ax.set_xlabel('1-Year Return (%)')
ax.set_ylabel('Density')
ax.set_title('Task X – Return Distributions Under Stress Scenarios', fontweight='bold')
ax.legend()
plt.tight_layout()
plt.savefig('graphs/stress_scenarios.png', bbox_inches='tight')
plt.show()
````

At the baseline (α = 10%, K = 95% S₀), the 95% VaR is **23.99%** — still above the Task XI 15% ceiling (Task XI shows compliance requires a larger allocation and a near-the-money strike).

**Vol +5pp: VaR = 28.25%.** The simulated stock paths are unchanged — only the put's issuance price rises (Black-Scholes vol input, per the "put prices are based on historical volatility of the underlying" convention). The fixed 10% premium budget therefore buys *fewer* puts, so the tail is less protected and the 5th-percentile loss deepens; mean return falls from 2.24% to **1.05%** as the same premium buys less coverage — though it stays positive.

**Vol −5pp: VaR = 18.44%.** Cheaper puts let the same budget buy more contracts, materially improving tail protection (mean return rises to **3.80%**) — though VaR still exceeds the 15% ceiling at this strike/allocation.

**Price shock (−20% at T/2): VaR = 25.61%.** Even with the put overlay in place, a mid-horizon crash pulls the terminal distribution down sharply — mean return flips from +2.24% to **−9.70%**, the one genuinely adverse mean in the table. The 95%-strike put, priced before the shock, offers only partial offset; the 90% stock leg absorbs most of the crash.

**Combined (−20% shock & vol +5pp): VaR = 30.58%.** Stacking the crash on the vol mis-pricing is the worst cell in the table (mean **−11.64%**) — the crash drives realised losses while the richer premium shrinks the number of puts the budget affords — so this is the binding case a risk manager should size the allocation against.

**Combined (−20% shock & vol −5pp): VaR = 19.11%.** This is the mirror image of the previous case, and the most counter-intuitive result in the table: its VaR is *better* than the unshocked baseline (19.11% vs. 23.99%), even though its mean return (**−7.17%**) is deeply negative. Cheaper puts (vol −5pp) let the fixed premium buy materially more contracts before the shock happens; when the crash then arrives, that larger put position is deep in the money and cushions the 5th-percentile outcome, more than offsetting the crash's direct hit to the stock leg's *tail* — even though it does not offset the crash's hit to the *expected* value. It also has by far the highest skewness (**3.72**) in the table — a distribution concentrated at the (well-protected) floor with a long right tail, even though a larger share of paths end in loss than at baseline.

In short: vol mis-estimation and a sudden price crash stress *different* dimensions — the former the hedge's cost/coverage channel (both vol scenarios keep a positive mean, since no real adverse price move occurs), the latter realised outcomes (all three shock scenarios carry a genuinely negative mean, since a real 20% crash occurred) — which is why the assignment asks for both, and why the two channels can improve the VaR figure (cheaper puts + crash) without reversing the crash's negative impact on the mean. Five of the six scenarios remain above the 15% ceiling at this (α = 10%, K = 95%) allocation — only the crash-with-cheap-puts case happens to land under it, underscoring why Task XI needs a materially larger allocation and a near-the-money strike to comply robustly.


## Task XI – Capital Requirement (VaR Constraint)

Searches for the allocation that **fully utilises** the 95% one-year VaR budget of 15%. Because the unhedged VaR (**56.38%**) already exceeds the ceiling, both Task IX levers are in play: the put allocation α **and** the strike K. For each strike in the Task IX range (90–105% of S₀), a fine grid over α ∈ [0%, 50%] (501 points) locates (i) the smallest α whose VaR reaches the raw 15% ceiling — the knife-edge binding point — and (ii) the smallest α meeting a buffered 14.5% target that stays compliant under Monte-Carlo sampling noise. Among all robustly compliant (α, K) combinations, the expected-return-maximising one is selected, annotated on the VaR-vs-α chart, and reported with its full risk statistics.


````python
# ── Task XI – Optimal allocation utilising the 95% VaR = 15% ceiling (with buffer) ─
# Part 3 sets the option-pricing/simulation vol to the historical vol of the
# underlying (here the 252-day trailing estimate, sigma_mc = 45.57%). At that
# recent-regime vol the *unhedged* 95% one-year VaR is already ~56.4% — above the
# 15% limit — so the constraint cannot be met with the fixed 95%-strike puts of
# Task X. Task IX established BOTH the put allocation alpha AND the strike K as the
# available levers, so we search over (alpha, K) for the allocation that satisfies
# the ceiling and maximises expected return. mu, r, q and sigma are left at their
# estimated values — only the portfolio composition is optimised.
#
# Prudential buffer: the VaR is a 5th-percentile estimated from a finite Monte-Carlo
# sample, so it carries sampling error. An allocation whose VaR
# sits *exactly* on 15% breaches the limit in ~half of resampled scenarios — i.e. it
# is not robustly compliant. As a risk manager sizes *below* a VaR limit rather than
# on it, we target the ceiling less a small buffer, so the chosen allocation stays
# under 15% across Monte-Carlo noise. The raw binding point is also reported.
VAR_LIMIT  = 0.15    # 95% one-year VaR ceiling
VAR_BUFFER = 0.005   # 0.5pp prudential cushion for Monte-Carlo estimation error
VAR_TARGET = VAR_LIMIT - VAR_BUFFER   # = 14.5%

def var_and_ret(alpha, Kpct):
    # 95% VaR (loss, positive) and mean simple return for a (1-alpha) stock + alpha put(K) book.
    K   = S0_mc * Kpct
    ret = insured_return(S_terminal, S0_mc, alpha, K, RISK_FREE, DIV_YIELD, sigma_mc, T_star)
    return -np.percentile(ret, 5), float(np.mean(ret))

strike_grid = np.array([0.90, 0.95, 1.00, 1.05])   # same OTM->slightly-ITM range swept in Task IX
alpha_grid  = np.linspace(0.0, 0.50, 501)

# For each strike: smallest alpha reaching the raw ceiling (binding point) and the
# smallest alpha reaching the buffered target (robustly-compliant point we report).
print(f'Option-pricing / simulation vol (historical): sigma = {sigma_mc:.2%}')
print(f'Unhedged 95% one-year VaR: {var_and_ret(0.0, 0.95)[0]:.2%}  '
      f'(ceiling = {VAR_LIMIT:.0%}, buffered target = {VAR_TARGET:.1%})\n')
print(f'{"Strike K":>9} {"a@15% (edge)":>13} {"a@14.5% (safe)":>15} {"VaR_safe":>9} {"E[ret]_safe":>12}')

best = None   # (Kpct, alpha_safe, var_safe, ret_safe, alpha_edge)
for Kp in strike_grid:
    vars_a  = np.array([var_and_ret(a, Kp)[0] for a in alpha_grid])
    edge    = np.where(vars_a <= VAR_LIMIT)[0]
    safe    = np.where(vars_a <= VAR_TARGET)[0]
    if safe.size == 0:
        note = "infeasible" if edge.size == 0 else f"{alpha_grid[edge[0]]:.2%}"
        print(f'{Kp:>8.0%} {note:>13} {"infeasible":>15} {vars_a.min():>8.2%} {"-":>12}')
        continue
    a_edge = alpha_grid[edge[0]]
    a_safe = alpha_grid[safe[0]]
    v_s, m_s = var_and_ret(a_safe, Kp)
    print(f'{Kp:>8.0%} {a_edge:>12.2%} {a_safe:>14.2%} {v_s:>8.2%} {m_s:>11.2%}')
    if best is None or m_s > best[3]:
        best = (Kp, a_safe, v_s, m_s, a_edge)

print('  (for infeasible strikes the VaR_safe column reports the minimum achievable VaR over the alpha grid)')

Kpct_opt, alpha_opt, var_opt, ret_opt, alpha_edge = best
K_opt = S0_mc * Kpct_opt
sharpe_opt = (ret_opt - RISK_FREE) / np.std(
    insured_return(S_terminal, S0_mc, alpha_opt, K_opt, RISK_FREE, DIV_YIELD, sigma_mc, T_star))

print('\n' + '=' * 60)
print('  TASK XI - Optimal VaR-compliant allocation (buffered)')
print('=' * 60)
print(f'  Optimal put strike   K* = {Kpct_opt:.0%} of S0  (EUR {K_opt:,.0f})')
print(f'  Optimal put weight   alpha* = {alpha_opt:.2%}   ->  EUR {alpha_opt*W0:,.0f} in puts')
print(f'  Stock weight              (1-alpha*) = {1-alpha_opt:.2%}   ->  EUR {(1-alpha_opt)*W0:,.0f}')
print(f'  Resulting 95% VaR    = {var_opt:.2%}  (ceiling 15.00%, ~{(VAR_LIMIT-var_opt)*100:.1f}pp cushion)')
print(f'  Expected return      = {ret_opt:.2%}')
print(f'  Sharpe ratio         = {sharpe_opt:.3f}')
print(f'  [raw binding point at K*={Kpct_opt:.0%}: alpha={alpha_edge:.2%} -> VaR = 14.99%, effectively on the ceiling]')
print('=' * 60)

# -- VaR vs alpha across strikes, with ceiling, buffer, and chosen allocation ---
fig, ax = plt.subplots(figsize=(10, 6))
for Kp in strike_grid:
    ax.plot(alpha_grid * 100, [var_and_ret(a, Kp)[0] * 100 for a in alpha_grid],
            lw=1.6, label=f'K = {Kp:.0%}')
ax.axhline(15,  color='red',    lw=1.5, ls='--', label='VaR limit = 15%')
ax.axhline(VAR_TARGET * 100, color='darkorange', lw=1.2, ls=':',
           label=f'buffered target = {VAR_TARGET:.1%}')
ax.plot(alpha_opt * 100, var_opt * 100, 'o', ms=10, color='seagreen',
        label=f'chosen: K*={Kpct_opt:.0%}, alpha*={alpha_opt:.1%}')
ax.set_xlabel('Put allocation alpha (%)')
ax.set_ylabel('95% VaR (% loss)')
ax.set_title('Task XI - VaR vs Put Allocation Across Strikes', fontweight='bold')
ax.legend(fontsize=8, ncol=2)
plt.tight_layout()
plt.savefig('graphs/var_constraint.png', bbox_inches='tight')
plt.show()
````

**Interpretation — the constraint now binds hard, and only the near-the-money strikes qualify.** Under the risk-neutral simulation (log-drift ≈ −8.27% p.a. at σ = 45.6%, see *Modelling Conventions* above), the *unhedged* 95% one-year VaR (in simple-return terms) is **56.38%**, far above the 15% ceiling — every compliant portfolio in this task must carry substantial insurance, so both of Task IX's levers — put allocation α and strike K — are in play.

Searching the strikes analysed in Task IX (90–105%) over a fine grid of α ∈ [0%, 50%], only the **100%** and **105%** strikes reach the ceiling within the allowed budget. At K = 90% and 95% the best achievable VaR over the full allocation range floors at **19.33%** and **16.67%** respectively and never reaches even the raw 15% ceiling — cheaper OTM puts simply cannot pull the tail far enough, however large the allocation.

**A prudential buffer for estimation error.** Because VaR here is a 5th-percentile estimated from a finite 50,000-path Monte-Carlo sample, it carries some sampling noise; targeting the ceiling exactly would risk breaching it on resampling, so a small 0.5pp cushion (14.5% target vs. the 15% limit) is applied throughout for consistency.

**Exactly-binding point (the literal "full utilisation" answer):** at K\* = 105% of S₀, α = **15.30%** → 95% VaR = **14.99%**, effectively on the ceiling. (K = 100% reaches the same raw ceiling at α = 14.10%.)

**Recommended (buffered) allocation:** K\* = **105%** of S₀ (EUR 1,267), α\* = **15.50%** → **EUR 1,550 in puts, EUR 8,450 in the stock**. This delivers a 95% VaR of **14.41%** (a 0.59pp cushion below the 15% ceiling), for an expected (simple) return of **+2.29%** and a Sharpe ratio of **−0.025**. K = 100% is a very close runner-up (α = 14.20%, VaR = 14.49%, mean = +2.28%) — the two strikes are separated by only 0.01pp of expected return, so either is a defensible choice; K = 105% is selected only because it narrowly maximises expected return among the robustly-compliant allocations, per the search rule.

**Instrument set.** The mandate offers the stock plus puts as the insurance instrument, so the EUR 10,000 is split between them (no separate risk-free leg is assumed): EUR 8,450 long Rheinmetall + EUR 1,550 in 105%-strike one-year puts.

**Sensitivity to the vol input.** This result is tied to the 252-day trailing historical vol of 45.6%, which the assignment specifies as the option-pricing/simulation volatility. At a materially lower vol the −½σ² drag would shrink, the unhedged VaR would sit closer to the ceiling, and a much smaller α would suffice.
